##graphs


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
car_crashes = sns.load_dataset('car_crashes')
car_crashes.head
```

```python
car_crashes.columns
```

```python
sns.histplot(x='speeding',data = car_crashes, bins=10, kde= True)
plt.show()
```

```python
car_crashes.columns
```

```python

sns.histplot(x= 'alcohol', data= car_crashes)
plt.title("car_crashes ka graph")
plt.xlabel("value")
plt.ylabel("you")
```
